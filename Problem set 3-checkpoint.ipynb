Step 2. Import the dataset from the attached file wind.txt

Step 3. Assign it to a variable called data and replace the first 3 columns by a properdatetime index.

```python

# parse_dates gets 0, 1, 2 columns and parses them as the index
data_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data'
data = pd.read_csv(data_url, sep = "\s+", parse_dates = [[0,1,2]]) 
data.head()
```

Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python

def fix_century(x):
  year = x.year - 100 if x.year > 1989 else x.year
  return datetime.date(year, x.month, x.day)

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_century)

data.head()
```

#  <b>Question 6</b>

Create a line plot showing the number of marriages and divorces per capita in theU.S. between 1867 and 2014. Label both lines and show the legend.Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
us_marriage_divorce_data = pd.read_csv('C:\\Users\\KAMALJIT\\Downloads\\us-marriages-divorces-1867-2014.csv')

# Extract columns
years = us_marriage_divorce_data['Year'].values
marriages_per_capita = us_marriage_divorce_data['Marriages_per_1000'].values
divorces_per_capita = us_marriage_divorce_data['Divorces_per_1000'].values

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(years, marriages_per_capita, label='Marriages per 1000', linewidth=2)
plt.plot(years, divorces_per_capita, label='Divorces per 1000', linewidth=2)

# Labeling and adding legend
plt.title('Marriages and Divorces per Capita in the U.S. (1867-2014)')
plt.xlabel('Year')
plt.ylabel('Per 1000 People')
plt.legend()

# Show the plot
plt.show()
```

# <b>Question 7</b>

Create a vertical bar chart comparing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000.

Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
us_marriage_divorce_data = pd.read_csv('C:\\Users\\KAMALJIT\\Downloads\\us-marriages-divorces-1867-2014.csv')

# Extract relevant data for the years 1900, 1950, and 2000
selected_years = [1900, 1950, 2000]
selected_data = us_marriage_divorce_data[us_marriage_divorce_data['Year'].isin(selected_years)]
marriages_per_capita = selected_data['Marriages_per_1000'].values
divorces_per_capita = selected_data['Divorces_per_1000'].values

# Plot the vertical bar chart
plt.figure(figsize=(10, 6))
bar_width = 0.35
bar_positions = range(len(selected_years))

plt.bar(bar_positions, marriages_per_capita, width=bar_width, label='Marriages per 1000')
plt.bar([pos + bar_width for pos in bar_positions], divorces_per_capita, width=bar_width, label='Divorces per 1000')

# Labeling
plt.title('Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')
plt.xlabel('Year')
plt.ylabel('Per 1000 People')
plt.xticks([pos + bar_width/2 for pos in bar_positions], selected_years)
plt.legend()

# Show the plot
plt.show()
```

# <b>Question 8</b>

Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort the actors by their kill count and label each bar with the corresponding actor's name.

Don't forget to label your axes!

For reference, here's the raw data:

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
actor_kill_data = pd.read_csv('C:\\Users\\KAMALJIT\\Downloads\\actor_kill_counts.csv')

# Sort the data by kill count
sorted_actor_data = actor_kill_data.sort_values(by='Count', ascending=True)

# Plot the horizontal bar chart
plt.figure(figsize=(10, 8))
plt.barh(sorted_actor_data['Actor'], sorted_actor_data['Count'], color='skyblue')
plt.xlabel('Kill Count')
plt.ylabel('Actor')
plt.title('Deadliest Actors in Hollywood')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Annotate each bar with the actor's name
for i, value in enumerate(sorted_actor_data['Count']):
    plt.text(value, i, str(value), ha='left', va='center')

# Show the plot
plt.show()
```

#  <b>Question 9</b>

Create a pie chart showing the fraction of all Roman Emperors that were assassinated.

Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
roman_emperors_data = pd.read_csv('C:\\Users\\KAMALJIT\\Downloads\\roman-emperor-reigns.csv')

# Count the number of emperors who were assassinated
assassinated_emperors = roman_emperors_data[roman_emperors_data['Cause_of_Death'] == 'Assassinated']
assassination_count = len(assassinated_emperors)

# Count the number of emperors who were not assassinated
non_assassinated_count = len(roman_emperors_data) - assassination_count

# Create data for the pie chart
labels = ['Assassinated', 'Not Assassinated']
sizes = [assassination_count, non_assassinated_count]
colors = ['#ff9999', '#66b3ff']
explode = (0.1, 0)  # explode the 1st slice (i.e., 'Assassinated')

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Fraction of Roman Emperors Assassinated')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle

# Show the plot
plt.show()
```
